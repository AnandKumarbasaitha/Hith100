```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.kdeplot(tips.tip)
```

```python
sns.kdeplot(tips.tip,shade=True)
```

```python
#Multivariate
sns.kdeplot(tips.tip,tips.total_bill)
```

```python
sns.kdeplot(tips.tip,tips.total_bill,shade = True)
```

```python
sns.kdeplot(tips.tip,tips.total_bill,n_levels=5)
```

```python
sns.kdeplot(tips.tip,tips.total_bill,n_levels=20)
```

```python
sns.kdeplot(tips.total_bill,kernel='gau')
sns.kdeplot(tips.total_bill,kernel='cos')
sns.kdeplot(tips.total_bill,kernel='biw')
sns.kdeplot(tips.tip,kernel='epa')
sns.kdeplot(tips.tip,kernel='tri')
```
