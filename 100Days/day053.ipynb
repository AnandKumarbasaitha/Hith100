```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.linspace(0,5,21)
print(x)
data = np.sin(x) + np.random.rand(5,21)
print(data)
```

```python
sns.tsplot(data)
```

```python
x = np.linspace(0,50,21)
# print(x)
data = np.sin(x) + np.random.rand(5,21)
# print(data)
```

```python
sns.tsplot(data)
```

```python
gammas = sns.load_dataset('gammas')
gammas.head()
```

```python
sns.tsplot(data=gammas, time='timepoint', value='BOLD signal',condition='ROI',unit='subject')
```

```python
sns.tsplot(data=data,err_style='ci_bars')
```

```python
sns.tsplot(data=data,err_style='ci_band')
```

```python
sns.tsplot(data=data,err_style='boot_traces')
```

```python
sns.tsplot(data=data,err_style='boot_kde')
```

```python
sns.tsplot(data=data,err_style='unit_traces')
```

```python
sns.tsplot(data=data,err_style='unit_points')
```

```python
sns.tsplot(data=data,err_style='ci_bars',interpolate=False)
```

```python
sns.tsplot(data=data,ci=[60,90])
```

```python
sns.tsplot(data=data,estimator=np.median)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxenplot(data=tips,x='tip')
```

```python
sns.boxenplot(data=tips,x='total_bill')
```

```python
sns.boxenplot(data=tips,x='day',y='tip')
```

```python
sns.boxenplot(data=tips,x='day',y='tip',hue='smoker')
```

```python
sns.boxenplot(data=tips,x='day',y='tip',hue='sex')
```

```python
sns.boxenplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.FacetGrid(tips,col='smoker',row='sex')
```

```python
sns.FacetGrid(tips,col='day',row='sex')
```

```python
g = sns.FacetGrid(tips,col='day',row='sex')
g.map(sns.distplot,'tip')
```

```python
g = sns.FacetGrid(tips,col='day',row='sex')
g.map(sns.kdeplot,'tip')
```

```python
g = sns.FacetGrid(tips,col='day',row='sex')
g.map(sns.boxplot,'tip')
```

```python
g = sns.FacetGrid(tips,col='day',row='sex')
g.map(sns.scatterplot,'tip','total_bill')
```
