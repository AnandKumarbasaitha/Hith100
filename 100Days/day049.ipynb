```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.violinplot('tip',data=tips)
```

```python
sns.violinplot('tip',data=tips, inner='box')
```

```python
sns.violinplot('tip',data=tips, inner='quartile')
```

```python
sns.violinplot('tip',data=tips, inner='stick')
```

```python
sns.violinplot('tip',data=tips, inner='point')
```

```python
sns.violinplot(y='tip',x='time',data=tips)
```

```python
sns.violinplot(y='tip',x='time',data=tips,scale='area')
```

```python
sns.violinplot(y='tip',x='time',data=tips,scale='count')
```

```python
sns.violinplot(y='tip',x='time',data=tips,scale='width')
```

```python
sns.violinplot(y='tip',x='time',data=tips,scale='count',hue='sex')
```

```python
sns.violinplot(y='tip',x='time',data=tips,scale='count',hue='sex',split=True)
```
