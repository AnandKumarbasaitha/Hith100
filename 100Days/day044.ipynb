```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

DistPlot

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.distplot(tips.tip)
```

```python
sns.distplot(tips.tip,hist=False)
```

```python
sns.distplot(tips.tip,rug=True)
```

```python
sns.distplot(tips.tip,kde=False)
```

```python
sns.distplot(tips.tip,kde=False)
```

```python
sns.distplot(tips.tip,kde=False,norm_hist = False)
```

```python
from scipy import stats
sns.distplot(tips.tip,fit=stats.gamma,kde = False)
```
