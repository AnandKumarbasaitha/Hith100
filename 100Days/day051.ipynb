```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pointplot(data=tips, x='day',y='tip')
```

```python
sns.pointplot(data=tips, x='day',y='tip',hue='smoker')
```

```python
sns.pointplot(data=tips, x='day',y='tip',hue='smoker',dodge = True)
```

```python
sns.pointplot(data=tips, x='day',y='tip',hue='smoker',markers=['o','x'],linestyles=['-','--'])
```

```python
sns.pointplot(data=tips, x='day',y='tip',hue='smoker',markers=['o','x'],linestyles=['-','--'],dodge=True,palette='Set1')
```

```python
sns.pointplot(data=tips, x='day',y='tip',join=False)
```

```python
sns.pointplot(data=tips, x='day',y='tip', estimator=np.mean)
```

```python
sns.pointplot(data=tips, x='day',y='tip', estimator=np.median)
```

```python
sns.pointplot(x="day", y="tip", data=tips, ci=50)
```

```python
sns.pointplot(x="day", y="tip", data=tips, capsize=1)
```

```python
sns.pointplot(x="day", y="tip", data=tips, capsize=0.05)
```

```python
sns.pointplot(x="day", y="tip", data=tips, capsize=0.1)
```

```python
sns.barplot(data=tips, x='day', y='tip')
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='time')
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='time',ci=0)
```

```python
sns.countplot(data=tips, x='day', hue='smoker')
```

```python
sns.countplot(data=tips, x='day', hue='sex')
```

```python
sns.countplot(data=tips, x='day', hue='smoker')
```
