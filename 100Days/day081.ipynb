```python
import nltk
import random
from nltk.corpus import pros_cons, movie_reviews
```

```python
corpus = movie_reviews
print(corpus.categories())
fileid = corpus.fileids('pos')
print(corpus.words(fileid))
```

```python
documents= []
for category in corpus.categories():
    for fileid in corpus.fileids(category):
        documents.append((list(corpus.words(fileid)),category))
```

```python
print(documents[5][0][:50])
print(documents[5][1])
print('\n')
print(documents[1200][0][:50])
print(documents[1200][1])
```

```python
words = []
for w in corpus.words():
    words.append(w.lower())
words
```

```python
words = nltk.FreqDist(words)
print(words.most_common(50))
print('\n',words["disgusting"])
print('\n',words["legendary"])
```

```python
word_features = list(words.keys())[:5000]
word_features
```

```python
def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
find_features(corpus.words('neg/cv000_29416.txt'))
```

```python
featuresets = [(find_features(review),category) for (review,category) in documents]
```

```python
print(len(featuresets))
train_set = featuresets[:1600]
test_set = featuresets[1600:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier,test_set)*100,'%')
print(classifier.show_most_informative_features(50))
```
