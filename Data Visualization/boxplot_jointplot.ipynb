```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

boxplot

```python
sns.boxplot('tip',data=tips)
```

```python
sns.boxplot('tip',data=tips,notch=True)
```

```python
sns.boxplot('tip',data=tips,whis=2)
```

```python
sns.boxplot('tip',data=tips,whis=0.5)
```

```python
sns.boxplot(y='tip',x='day',data=tips)
```

```python
sns.boxplot(y='tip',x='day',hue='smoker',data=tips)
```

```python
sns.boxplot(y='tip',x='day',data=tips,palette='Set1')
```

```python
sns.boxplot(y='tip',x='day',data=tips,palette='Set2')
```

```python
sns.boxplot(y='tip',x='day',data=tips,palette='Set3')
```

```python
sns.boxplot(y='tip',x='day',data=tips)
sns.swarmplot(y='tip',x='day',data=tips,color='Black')
```

jointplot

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde')
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',stat_func=spearmanr)
```

```python
sns.jointplot(data=tips,x='size',y='tip',stat_func=spearmanr)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips).plot_joint(sns.kdeplot,n_levels=4)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips).plot_joint(sns.kdeplot,n_levels=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips).plot_joint(sns.kdeplot,n_levels=2)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips).plot_joint(sns.kdeplot,n_levels=2).plot_marginals(sns.rugplot)
```
