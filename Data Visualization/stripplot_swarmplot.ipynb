```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.stripplot(tips['tip'])
```

```python
sns.stripplot(x='time',y='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=1)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=0.05)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=0.5)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=0.5)
```

```python
sns.stripplot(y='day',x='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',hue='time',data=tips)
```

```python
sns.stripplot(x='day',y='tip',hue='sex',data=tips)
```

```python
#Split parameter is renamed to dodge
sns.stripplot(x='day',y='tip',hue='sex',data=tips,dodge=True)
```

```python
sns.stripplot(x='day',y='tip',hue='time',data=tips,dodge=True)
```

```python
sns.swarmplot(x='day',y='tip',hue='sex',data=tips)
```
