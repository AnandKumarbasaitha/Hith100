```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Called Catplot in newer version
sns.factorplot(data=tips, x='day',y='tip')
```

```python
# default kind is strip instead of point
sns.catplot(data=tips, x='day',y='tip')
```

```python
sns.catplot(data=tips, x='day',y='tip',kind='point')
```

```python
sns.catplot(data=tips, x='day',y='tip',kind='violin')
```

```python
sns.catplot(data=tips, x='day',y='tip',hue='sex',dodge=True)
```

```python
sns.catplot(data=tips, x='day',y='tip',hue='sex',dodge=True,col='smoker',row='time',kind='point')
```

```python
sns.catplot(data=tips, x='day',y='tip',hue='sex',dodge=True, kind='violin')
```

```python
sns.catplot(data=tips, x='day',y='tip',hue='sex',dodge=True, kind='bar')
```

```python
sns.catplot(data=tips, x='day',y='tip',hue='sex',dodge=True,col='smoker',row='time',kind='box')
```

```python
sns.catplot(data=tips, x='tip',y='time',hue='sex',dodge=True,row='smoker',col='day',kind='box')
```

```python
uniform_data = np.random.uniform(0,10,(7,10))
uniform_data
```

```python
sns.heatmap(uniform_data)
```

```python
sns.heatmap(uniform_data,annot=True)
```

```python
sns.heatmap(uniform_data,vmin=1,vmax=7)
```

```python
sns.heatmap(uniform_data,robust=True)
```

```python
sns.heatmap(uniform_data,annot=True,robust=True)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month','year','passengers')
```

```python
flights
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,robust=True)
```

```python
normal_data = np.random.randn(12,8)
print(normal_data.shape)
normal_data
```

```python
sns.heatmap(normal_data)
```

```python
sns.heatmap(normal_data,robust=True)
```

```python
sns.heatmap(normal_data,xticklabels=4,yticklabels=3)
```

```python
sns.heatmap(normal_data,square=True)
```

```python
sns.heatmap(normal_data,xticklabels=4,yticklabels=3,square=True)
```

```python
normal_data2=np.random.randn(4,5)
sns.heatmap(normal_data2,annot=True)
corr = np.corrcoef(normal_data2)
print(corr.shape)
print(corr)
```

```python
sns.heatmap(corr,annot=True)
```

```python
mask = np.zeros_like(corr)
print(mask,'\n')
mask[np.triu_indices_from(mask)] = True
print(mask)
```

```python
sns.heatmap(corr,mask=mask,annot=True)
```
