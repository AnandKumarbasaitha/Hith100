```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
anscombe = sns.load_dataset("anscombe")
print(anscombe.head())
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='smoker')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='time')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='day')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='size')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='size',truncate=True)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='smoker')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',row='sex',col='smoker')
```

```python
sns.lmplot(data=tips,x='size',y='tip')
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_jitter = 0.1)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.mean)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.median)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.mean, ci=90)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.mean,x_ci=99.9999)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.mean,x_ci=90)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.mean,x_ci=75)
```

```python
sns.lmplot(data=tips,x='size',y='tip',x_estimator = np.mean,x_ci=40)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',x_bins=100)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',x_bins=10)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',x_bins=5)
```

```python
sns.lmplot(data=anscombe.query("dataset=='I'"),x='x',y='y')
```

```python
sns.lmplot(data=anscombe.query("dataset=='I'"),x='x',y='y',ci=75)
```

```python
sns.lmplot(data=anscombe.query("dataset=='I'"),x='x',y='y',ci=None)
```

```python
sns.lmplot(data=anscombe.query("dataset=='II'"),x='x',y='y',ci=None)
```

```python
sns.lmplot(data=anscombe.query("dataset=='II'"),x='x',y='y',ci=None,order=2)
```

```python
sns.lmplot(data=anscombe.query("dataset=='III'"),x='x',y='y',ci=None,order=1)
```

```python
sns.lmplot(data=anscombe.query("dataset=='III'"),x='x',y='y',ci=None,order=2)
```

```python
#An n-1 degree polynomial can fit n data points
sns.lmplot(data=anscombe.query("dataset=='III'"),x='x',y='y',ci=None,order=9)
```

```python
#But the point corresponding to x=13 seems to be an outlier
sns.lmplot(data=anscombe.query("dataset=='III'"),x='x',y='y',robust=True,ci=None)
```

```python
#People who give integer amounts as tips
tips['big_tip'] = (tips.tip/tips.total_bill)>0.15
tips.head()
```

```python
sns.lmplot(data=tips,x='total_bill',y='big_tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='big_tip',y_jitter=0.05)
```

```python
sns.lmplot(data=tips,x='total_bill',y='big_tip',y_jitter=0.05,logistic=True)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',y_jitter=0.05,lowess=True)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',y_jitter=0.05,logx =True)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',y_jitter=0.05,logx =True,truncate=True)
```

```python
sns.lmplot(data=tips,x='total_bill',y='big_tip',y_jitter=0.05,logistic =True)
```
